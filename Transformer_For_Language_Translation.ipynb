```python
from transformers import MarianMTModel, MarianTokenizer

# Define models for each direction
MODELS = {
    'hi-en': 'Helsinki-NLP/opus-mt-hi-en',
    'en-hi': 'Helsinki-NLP/opus-mt-en-hi'
}

def load_model(direction='hi-en'):
    model_name = MODELS[direction]
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model = MarianMTModel.from_pretrained(model_name)
    return tokenizer, model

def translate(text, direction='hi-en'):
    tokenizer, model = load_model(direction)
    # Tokenize and prepare inputs
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    # Generate translation
    translated = model.generate(**inputs)
    # Decode outputs
    return tokenizer.decode(translated[0], skip_special_tokens=True)

# Example usage
hindi_text = "मुझे अंग्रेजी बोलना पसंद है।"
english_text = "I like to speak Hindi."

#hindi_text = "प्रकृति हवा में फुसफुसाती है, हर पत्ते में जीवन का संगीत गूंजता है। इसकी सुंदरता हर सूर्योदय, हर लहर, और हर ऊँचा खड़ा पेड़ में संतुलन में निहित है।"
#english_text = "Nature whispers in the breeze, a symphony of life unfolding in every leaf. Its beauty lies in the balance of every sunrise, every ripple, and every tree standing tall."

# Hindi to English
translated_en = translate(hindi_text, direction='hi-en')
print("Hindi to English:", translated_en)

# English to Hindi
translated_hi = translate(english_text, direction='en-hi')
print("English to Hindi:", translated_hi)
```
